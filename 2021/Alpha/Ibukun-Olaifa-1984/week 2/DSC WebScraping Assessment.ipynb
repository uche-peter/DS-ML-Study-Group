# APPROACH TO GETTING ALL BOOKS DETAILS

As we all know there various ways to scrape books from websites, Well for me i'll be scraping each pages of the website and printing out the book names, price, image_urls and ratings.

# SCRAPING

After inspecting the website i figure out each page has a unique link, so i decided to get each pages of websites, since each pages has books in it

# LET'S GO
Here is the the trick we get all the pages of the website, Create an empty list for the prices, book name, image URLS and ratings for each book. Then from each page we scrape and append all the details to the empty list. Easy peasy right!😋

```python
url = 'http://books.toscrape.com/'

prices = []
book_names = []
image_urls = []
ratings = []

def page_links():
    for i in range(1, 51): #there are 50 pages on the website, as said earlier we get each pages from the site
        page_links = url + f"catalogue/page-{i}.html" # pages
        pages = requests.get(page_links)
        bs_page = bs(pages.content)
#         print(bs_page.prettify()) #prints each pages of the websites
#         print('\n')
#         print('-'*200)
        book_lists = bs_page.find_all('li', attrs={'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'}) #Lists of all books
    
        """Gets the price"""
        for price in book_lists:
            book_price = price.find('p', attrs={'class':'price_color'})
            prices.append(book_price.string)
        
        """Gets the book name"""

        for book in book_lists:
            book_name = book.find('h3').find('a')['title']
            book_names.append(book_name)
        
        """
        Gets the rating, This part is kind of tricky, when you inspect the p tag, u will definitely find out that there
        are two class, the first class represents the  paragraph star rating icon ("star-rating") while the second class 
        represents the actual rating, so we get the 2nd class since that is what we are looking for.

        """

        for rating in book_lists:
            rates = rating.find('p')['class'][1]
            ratings.append(rates)
        
        """Gets the image_urls"""

        for image_link in book_lists:
            image_url = image_link.select_one('div.image_container img')
            image_url = url + image_url['src'].strip('../')
            image_urls.append(image_url)
```

# PUTTING THE BOOK DETAILS TO  PANDAS DATAFRAME

```python
df = pd.DataFrame(data={'Book Name':book_names, 'Image URL':image_urls, 'Prices':prices, 'Ratings':ratings}, index=range(1,1001))
```
